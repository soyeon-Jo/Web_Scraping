# csv 기본
### 네이버 금융
시가총액 TOP 200 종목명, 현재가, 전일비 등을 가져와서 csv파일로 저장하기 

```python
import csv
import requests
from bs4 import BeautifulSoup

url = "https://finance.naver.com/sise/sise_market_sum.nhn?&page="

filename = "시가총액1-200.csv"
f = open(filename, "w", encoding ="utf-8-sig", newline="")
#open(파일이름, 쓰는 방식, 인코딩방법, newline="" )
#newline ="" <-줄바꿈
# 인코딩 방법 : utf -8 
# utf-8이용시 한글 깨짐 현상이 있다면 utf-8-sig로
writer = csv.writer(f) # 파일을 쓰기(write)

title = "N	종목명	현재가	전일비	등락률	액면가	시가총액	상장주식수	외국인비율	거래량	PER	ROE	토론실".split("\t")
writer.writerow(title) #writerow : row열 추가

for page in range(1,5):
    res = requests.get(url + str(page))
    res.raise_for_status()
    
    soup = BeautifulSoup(res.text, "lxml")
    
    data_rows = soup.find("table", attrs={"class":"type_2"}).find("tbody").find_all("tr")
    
    for row in data_rows:
        columns = row.find_all("td")
        
        if len(columns) <= 1 :
            continue
        data = [column.get_text().strip() for column in columns ]
        print(data)

```
